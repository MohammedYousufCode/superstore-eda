This report summarizes key findings from the Superstore Sales EDA.
It covers 9,994 orders across 4 regions, 3 categories, and 17 sub-categories.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.float_format = '{:,.2f}'.format

sns.set_theme(style='whitegrid')

df = pd.read_csv('../data/Sample - Superstore.csv', encoding='latin-1')
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
df['order_year'] = df['order_date'].dt.year
df['order_month'] = df['order_date'].dt.month
df['shipping_days'] = (df['ship_date'] - df['order_date']).dt.days
```

## Executive Summary

The Superstore dataset covers 4 years of retail sales (2014–2017) with a total 
revenue of ~2.3M and total profit of ~286K — roughly a 12% overall margin.

Key finding: Revenue and profit are not evenly distributed. 
A small number of products, regions, and categories drive most of the profit, 
while certain combinations actively lose money.

```python
print("=== Overall Business Summary ===")
print(f"Total Orders      : {df.shape[0]:,}")
print(f"Total Sales       : ${df['sales'].sum():,.0f}")
print(f"Total Profit      : ${df['profit'].sum():,.0f}")
print(f"Overall Margin    : {(df['profit'].sum()/df['sales'].sum()*100):.1f}%")
print(f"Avg Discount      : {df['discount'].mean()*100:.2f}%")
print(f"Loss-making Orders: {(df['profit'] < 0).sum():,} ({(df['profit']<0).mean()*100:.1f}%)")
```

## 5 Key Business Findings

### Finding 1 — Technology is the most profitable category
Technology generates ~145K profit on ~836K sales (17% margin).
Furniture generates only ~18K profit on ~742K sales (2% margin).
Despite similar revenue, Furniture barely breaks even.

### Finding 2 — West region leads in profitability
West contributes ~108K profit, the highest of all 4 regions.
Central region is the weakest — it has decent sales but the lowest profit margin,
suggesting discount overuse or high operational cost.

### Finding 3 — Standard Class dominates shipping
60% of all orders use Standard Class shipping.
Same Day delivery accounts for less than 6% of orders.
This means customers strongly prefer cost over speed.

### Finding 4 — Heavy discounting directly destroys profit
Orders with 0% discount are almost always profitable.
Orders with 40%+ discount are consistently loss-making across all categories.
The data shows a clear inverse relationship between discount level and profit.

### Finding 5 — Tables and Bookcases are loss-making sub-categories
Tables have a total profit of -$17,725  despite generating   $206K in sales.
Bookcases have -$3,473 total profit.
These two sub-categories are dragging down the entire Furniture category.

## 3 Business Recommendations

### Recommendation 1 — Cap discounts at 20%
The data clearly shows profitability collapses above 30% discount.
Introducing a maximum discount policy of 20% would protect margins
across all categories, especially Office Supplies and Technology.

### Recommendation 2 — Review Furniture pricing strategy
Tables and Bookcases are selling but losing money.
Options: increase base price, reduce discount allowances,
or phase out the weakest SKUs from the product catalog.

### Recommendation 3 — Invest more in West and East regions
West and East contribute the highest profit per sales dollar.
Marketing and inventory investment in these two regions
would yield better returns compared to Central or South.
